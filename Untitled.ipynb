```python
import pandas as pd
```

```python
df = pd.read_csv("Data/cleanprep.csv")
df
```

## Quick Data Cleaning

```python
df.iloc[0, :] = df.columns
cols_to_add = ["DateTime", "Username", "Text", "Polarity", "Sensitivity"]
df.columns = cols_to_add
```

```python
drop_str_b = ["Username", "Text"]
for col in drop_str_b:
    df[col] = df[col].map(lambda x: x[2:len(x) -1])
df
```

```python
df["DateTime"] = pd.to_datetime(df["DateTime"])
df
```

```python
df.info()
```

```python
df["Polarity"] = df["Polarity"].astype("float64")
df["Sensitivity"] = df["Sensitivity"].astype("float64")
```
